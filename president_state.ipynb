```python
import pandas as pd
```

```python
def fips_code(x):
    l=[]
    for i in x:
        if i <10:
            l.append("04000US0"+str(i))
        else:
            l.append("04000US"+str(i))
    return l
```

```python

president=pd.read_csv("https://dataverse.harvard.edu/api/access/datafile/:persistentId?persistentId=doi:10.7910/DVN/42MVDX/MFU99O",delimiter='\t')

```
